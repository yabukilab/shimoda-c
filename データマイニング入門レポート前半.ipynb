####2.4 数値最適化(scipy.optimize)

**課題** $L(a,b):=a^2-4a+b^2-6b+18$を最小にするa,bを,数値最適化によって求めよ.

ヒント:$L(a,b)=(a-2)^2+(b-3)^2+5$である.

```python
def L(v):
  a, b = v
  return a**2 - 4 * a + b**2 - 6 * b + 18

minimize(L, (0, 0))
```

**課題** 例1のデータについて,数値最適化によって,aとbを求めよ.

```python
x = data1["年齢階級"]
y = data1["血圧の平均"]

def L(v):
  a, b = v
  return sum((y - a * x - b)**2)

minimize(L, (0, 0))
```

**課題** data2に対する,線形回帰分析とK最近傍法(K=5)の予測性能を,5分割交差検証で調べて,次のように可視化せよ.性能指標は決定係数(r2)とする.

```python
X = data2[["speed"]]
y = data2["dist"]

model1 = LinearRegression()
scores1 = cross_val_score(model1, X, y, cv=5, scoring='r2')

model2 = KNeighborsRegressor(n_neighbors=5)
scores2 = cross_val_score(model2, X, y, cv=5, scoring='r2')

pd.DataFrame({
    '線形回帰分析' : scores1,
    'K最近傍法' : scores2,
}).boxplot();
```

**課題** data2に対する,線形回帰分析とK最近傍法(K=5)の予測性能を,5分割交差検証で調べて,可視化せよ.性能指標はMSE(neg_mean_squared_error)とする.

```python
X = data2[["speed"]]
y = data2["dist"]

model1 = LinearRegression()
scores1 = cross_val_score(model1, X, y, cv=5, scoring='neg_mean_squared_error')

model2 = KNeighborsRegressor(n_neighbors=5)
scores2 = cross_val_score(model2, X, y, cv=5, scoring='neg_mean_squared_error')

pd.DataFrame({
    '線形回帰分析' : scores1,
    'K最近傍法' : scores2,
}).boxplot();
```

**課題** data2に対する,線形回帰分析とK最近傍法(K=5)の予測性能を,一つ抜き交差検証(leave-one-out cross-validation, LOOCV)で調べて,可視化せよ.性能指標はMSEとする.

```python
X = data2[["speed"]]
y = data2["dist"]

model1 = LinearRegression()
scores1 = cross_val_score(model1, X, y, cv=LeaveOneOut(), scoring='neg_mean_squared_error')

model2 = KNeighborsRegressor(n_neighbors=5)
scores2 = cross_val_score(model2, X, y, cv=LeaveOneOut(), scoring='neg_mean_squared_error')

pd.DataFrame({
    '線形回帰分析' : scores1,
    'K最近傍法' : scores2,
}).boxplot();
```
